### Blood Pressure column

```python
print('The unique values of Blood Pressure column are', sleep_data['Blood Pressure'].unique())
```

```python
Blood_Pressure_count=sleep_data['Blood Pressure'].value_counts().reset_index()
Blood_Pressure_count
```

#### Note üìã
#### üî¥ Ideal blood pressure systolic (upper number) : less than 120 , diastolic (bottom number) : less than 80

#### üî¥ Normal systolic (upper number) : in range (120 - 129) , diastolic (bottom number) : in range (80 - 84)

#### üî¥ Otherwise, blood pressure is high

```python
# Group the data by 'Occupation', 'Gender', and 'Blood Pressure', count occurrences, and reset the index
position_blood = sleep_data.groupby(['Occupation', 'Gender', 'Blood Pressure']).size().reset_index(name='count')
position_blood
```

#### Insightsüí°
#### üî∏ Nurses (Female): Have the highest count of elevated blood pressure readings, with 65 instances recorded at 140/95.
#### üî∏ Lawyers (Male): Show a significant number of high blood pressure readings, with 42 instances recorded at various levels, including 130/85 and 135/88.
#### üî∏ Engineer (Female): Most common blood pressure reading is 125/80, with 32 instances.
